```python
from pyrosm import OSM
import geopandas as gpd
```

```python
# Load the .osm.pbf file
osm = OSM(r"C:\Users\............\massachusetts.osm.pbf")

# Get only road edges
roads = osm.get_data_by_custom_criteria(
    custom_filter={"highway": True},  # Filter for road features
    filter_type="keep",               # Keep all attributes of roads
    keep_nodes=False                  # Don't include intersection points (nodes)
)
```

```python
# Inspect data
roads.head()
```

```python
# Save the shapefile in the Downloads folder
output_path = r"C:\Users\............\massachusetts_roads.shp"
roads.to_file(output_path)
```
